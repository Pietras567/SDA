Indeksy jakości powietrza dla poszczególnych zanieczyszczeń liczone są w oparciu o wartości 1-godzinnych stężeń tych zanieczyszczeń, przypisywanych do odpowiedniej kategorii z tabeli zakresów dla indeksów poszczególnych zanieczyszczeń. Następnie indeks ogólny przyjmuje wartość najgorszego indeksu indywidualnego spośród zanieczyszczeń mierzonych na tej stacji lub przyjmuje wartość zanieczyszczenia dominującego dla województwa (pył zawieszony lub ozon) lub indeks nie jest określany i wyświetlany jest w kolorze szarym. „Brak indeksu” wskazuje na to, iż na danej stacji nie prowadzi się automatycznych pomiarów zanieczyszczenia, które w danym czasie decyduje o jakości powietrza na obszarze województwa. W okresie jesienno-zimowym dotyczy to przeważnie pyłu zawieszonego PM2,5/PM10, a w okresie wiosenno-letnim – ozonu.

Zaś dla poszczególnych kategorii Polskiego Indeksu Jakości Powietrza wyróżnić możemy następujące informacje zdrowotne:<br>
**Bardzo dobry** - Jakość powietrza jest bardzo dobra, zanieczyszczenie powietrza nie stanowi zagrożenia dla zdrowia, warunki bardzo sprzyjające do wszelkich aktywności na wolnym powietrzu, bez ograniczeń.<br>
**Dobry** - Jakość powietrza jest zadowalająca, zanieczyszczenie powietrza powoduje brak lub niskie ryzyko zagrożenia dla zdrowia. Można przebywać na wolnym powietrzu i wykonywać dowolną aktywność, bez ograniczeń.<br>
**Umiarkowany** - Jakość powietrza jest akceptowalna. Zanieczyszczenie powietrza może stanowić zagrożenie dla zdrowia w szczególnych przypadkach (dla osób chorych, osób starszych, kobiet w ciąży oraz małych dzieci). Warunki umiarkowane do aktywności na wolnym powietrzu.<br>
**Dostateczny** - Jakość powietrza jest dostateczna, zanieczyszczenie powietrza stanowi zagrożenie dla zdrowia (szczególnie dla osób chorych, starszych, kobiet w ciąży oraz małych dzieci) oraz może mieć negatywne skutki zdrowotne. Należy rozważyć ograniczenie (skrócenie lub rozłożenie w czasie) aktywności na wolnym powietrzu, szczególnie jeśli ta aktywność wymaga długotrwałego lub wzmożonego wysiłku fizycznego.<br>
**Zły** - Jakość powietrza jest zła, osoby chore, starsze, kobiety w ciąży oraz małe dzieci powinny unikać przebywania na wolnym powietrzu. Pozostała populacja powinna ograniczyć do minimum wszelką aktywność fizyczną na wolnym powietrzu - szczególnie wymagającą długotrwałego lub wzmożonego wysiłku fizycznego.<br>
**Bardzo zły** - Jakość powietrza jest bardzo zła i ma negatywny wpływ na zdrowie. Osoby chore, starsze, kobiety w ciąży oraz małe dzieci powinny bezwzględnie unikać przebywania na wolnym powietrzu. Pozostała populacja powinna ograniczyć przebywanie na wolnym powietrzu do niezbędnego minimum. Wszelkie aktywności fizyczne na zewnątrz są odradzane. Długotrwała ekspozycja na działanie substancji znajdujących się w powietrzu zwiększa ryzyko wystąpienia zmian m.in. w układzie oddechowym, naczyniowo-sercowym oraz odpornościowym.<br>
**Brak indeksu** - „Brak Indeksu” odpowiada sytuacji, gdy na danej stacji pomiarowej nie są aktualnie prowadzone pomiary pyłu zawieszonego lub ozonu, a jeden z nich jest w danej chwili decydującym zanieczyszczeniem powietrza w województwie. Indeks Jakości Powietrza nie jest wtedy wyznaczany, a kolor punktów na mapie bieżących danych pomiarowych zmienia się na szary. # Mapa interaktywna oraz statyczna stacji pomiarowych

**Ten kod generuje interaktywną mapę za pomocą biblioteki Folium. Mapa wyświetla dane o stacjach monitorowania jakości powietrza w Polsce, kategoryzując je na aktywne (zielone markery) i nieaktywne (czerwone markery) na podstawie dat zamknięcia stacji. Tworzy podpowiedzi (tooltip) i szczegółowe okna (popup) dla każdej stacji, dodaje kontrolę warstw oraz legendę dla łatwiejszej interpretacji. Mapa jest zapisywana jako plik HTML i wyświetlana w notebooku.**

```python
# Przekształć kolumnę 'TerminatinDate' na format daty
station_data = gdf3.copy()
station_data['TerminatinDate'] = pd.to_datetime(station_data['TerminatinDate'], format="%d/%m/%Y", errors='coerce')

# Określ, które stacje są aktywne na koniec 2023 roku
end_date = pd.Timestamp("2023-12-31")
station_data['IsActive'] = station_data['TerminatinDate'].isna() | (station_data['TerminatinDate'] > end_date)

map_center = [51.77513554189464, 19.455341364192083]
station_map = folium.Map(location=map_center, zoom_start=6)

station_data['FormattedTerminationDate'] = station_data['TerminatinDate'].dt.strftime('%d/%m/%Y')
station_data['FormattedTerminationDate'] = station_data['FormattedTerminationDate'].fillna("Brak")

# Funkcja tworząca opis markera (pomijając OldStationName)
def create_popup(row):
    popup_content = f"""
    <table style="width: 300px;">
        <tr><th style="text-align: left;">Number</th><td>{row['Number']}</td></tr>
        <tr><th style="text-align: left;">Station ID</th><td>{row['StationID']}</td></tr>
        <tr><th style="text-align: left;">International Station ID</th><td>{row['InternationalStationID']}</td></tr>
        <tr><th style="text-align: left;">Station Name</th><td>{row['StationName']}</td></tr>
        <tr><th style="text-align: left;">Start Date</th><td>{row['StartDate']}</td></tr>
        <tr><th style="text-align: left;">Termination Date</th><td>{row['FormattedTerminationDate']}</td></tr>
        <tr><th style="text-align: left;">Type</th><td>{row['Type']}</td></tr>
        <tr><th style="text-align: left;">Area</th><td>{row['Area']}</td></tr>
        <tr><th style="text-align: left;">Station Type</th><td>{row['StationType']}</td></tr>
        <tr><th style="text-align: left;">State</th><td>{row['State']}</td></tr>
        <tr><th style="text-align: left;">City</th><td>{row['City']}</td></tr>
        <tr><th style="text-align: left;">Address</th><td>{row['Address']}</td></tr>
        <tr><th style="text-align: left;">Latitude</th><td>{row['lat']}</td></tr>
        <tr><th style="text-align: left;">Longitude</th><td>{row['long']}</td></tr>
    </table>
    """
    return popup_content

# Dodanie grup markerów dla aktywnych i nieaktywnych stacji
active_group = folium.FeatureGroup(name="Aktywne stacje", show=True, control=True)
inactive_group = folium.FeatureGroup(name="Nieaktywne stacje", show=True, control=True)

# Dodanie stacji do odpowiednich grup
for index, row in station_data.iterrows():
    lat, lon = row['lat'], row['long']
    popup = folium.Popup(create_popup(row), max_width=300)
    
    # Kolor znacznika w zależności od aktywności
    if row['IsActive']:
        color = "green"
        folium.Marker(
            location=[lat, lon],
            popup=popup,
            tooltip=row['StationName'],
            icon=folium.Icon(color=color, prefix="fa", icon="wind"),
        ).add_to(active_group)
    else:
        color = "red"
        folium.Marker(
            location=[lat, lon],
            popup=popup,
            tooltip=row['StationName'],
            icon=folium.Icon(color=color, prefix="fa", icon="wind"),
        ).add_to(inactive_group)

# Dodanie grup do mapy
active_group.add_to(station_map)
inactive_group.add_to(station_map)

# Dodanie legendy
legend_html = """
<div style="position: fixed; 
            bottom: 50px; left: 50px; width: 200px; height: 90px; 
            background-color: white; z-index:9999; font-size:14px;
            border:1px solid grey; border-radius:5px; padding:10px;">
    <b>Legenda</b><br>
    <i class="fa-solid fa-wind" style="color:green"></i> Aktywne stacje<br>
    <i class="fa-solid fa-wind" style="color:red"></i> Nieaktywne stacje
</div>
"""
station_map.get_root().html.add_child(folium.Element(legend_html))

# Dodanie przełącznika warstw
folium.LayerControl().add_to(station_map)

# Zapisanie mapy do pliku HTML
station_map.save("station_map_with_activity.html")

station_map
```

```python
mpl_style(True)

poland_border = gpd.read_file('data/poland_country.geojson')
cities_data = pd.read_csv('data/cities_data.csv', sep=',')

# Filtruj miasta z dużą lub średnią populacją (np. powyżej 200 tys.)
large_cities = cities_data[cities_data['population'] > 200000]

# Konwersja na GeoDataFrame
cities_gdf = gpd.GeoDataFrame(
    large_cities,
    geometry=[Point(xy) for xy in zip(large_cities['lng'], large_cities['lat'])],
    crs="EPSG:4326"  # WGS84
)

# Filtrujemy dane dla aktywnych stacji
active_stations = station_data[station_data['IsActive']].copy()

# Tworzymy punkty geometryczne z lat i long
active_stations['geometry'] = active_stations.apply(lambda row: Point(row['long'], row['lat']), axis=1)

# Tworzenie podstawowej mapy statycznej
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_title("Mapa aktywnych stacji na koniec 2023 roku", fontsize=16)
ax.set_xlabel("Długość geograficzna", fontsize=12)
ax.set_ylabel("Szerokość geograficzna", fontsize=12)

# Rysowanie granicy Polski
poland_border.plot(ax=ax, color='lightblue', edgecolor='black', alpha=0.5)

# Dodanie miast jako punkty
cities_gdf.plot(
    ax=ax,
    color='#FE0050',
    markersize=cities_gdf['population'] / 5000,  # Skalowanie wielkości punktów według populacji
    alpha=0.7
)

# Rysowanie aktywnych stacji na mapie
for _, station in active_stations.iterrows():
    ax.plot(
        station.geometry.x,  # Długość geograficzna (longitude)
        station.geometry.y,  # Szerokość geograficzna (latitude)
        'o',  # Styl punktu (okrąg)
        markersize=5,
        color='#5DFE9A',  # Kolor dla aktywnych stacji
        label='Aktywne stacje' if 'Aktywne stacje' not in ax.get_legend_handles_labels()[1] else "",  # Unikaj duplikacji legendy
        alpha=0.6
    )
    
# Dodanie nazw miast
for x, y, label in zip(cities_gdf.geometry.x, cities_gdf.geometry.y, cities_gdf['city']):
    ax.text(x, y, label, fontsize=8, ha='right', color='#4323FF', fontweight='bold', alpha=1)

# Dodanie legendy
ax.legend(loc='upper right', fontsize=10)

# Wyświetlenie mapy
plt.grid()
plt.show()

# Zapisanie mapy jako plik graficzny
fig.savefig("static_active_station_map.png", dpi=300, bbox_inches="tight")
```

**Kod definiuje pięć funkcji (`NO2`, `SO2`, `PM25`, `PM10` i `O3`), które klasyfikują wartości jakości powietrza na kategorie (-1 do 5) w oparciu o zdefiniowane zakresy stężeń zanieczyszczeń.**

```python
# -1 Brak indeksu
#  0 Bardzo dobry
#  1 Dobry
#  2 Umiarkowany
#  3 Dostateczny
#  4 Zły
#  5 Bardzo zły

def NO2(value):
    ranges = {
        (0, 40): 0,
        (40.1, 100): 1,
        (100.1, 150): 2,
        (150.1, 230): 3,
        (230.1, 400): 4,
        (400.1, float('inf')): 5,
    }
    for (low, high), label in ranges.items():
        if low < value <= high:
            return label
    return -1

def SO2(value):
    ranges = {
        (0, 50): 0,
        (50.1, 100): 1,
        (100.1, 200): 2,
        (200.1, 350): 3,
        (350.1, 800): 4,
        (800.1, float('inf')): 5,
    }
    for (low, high), label in ranges.items():
        if low < value <= high:
            return label
    return -1

def PM25(value):
    ranges = {
        (0, 13): 0,
        (13.1, 35): 1,
        (35.1, 55): 2,
        (55.1, 75): 3,
        (75.1, 110): 4,
        (110.1, float('inf')): 5,
    }
    for (low, high), label in ranges.items():
        if low < value <= high:
            return label
    return -1

def PM10(value):
    ranges = {
        (0, 20): 0,
        (20.1, 50): 1,
        (50.1, 80): 2,
        (80.1, 110): 3,
        (110.1, 150): 4,
        (150.1, float('inf')): 5,
    }
    for (low, high), label in ranges.items():
        if low < value <= high:
            return label
    return -1

def O3(value):
    ranges = {
        (0, 70): 0,
        (70.1, 120): 1,
        (120.1, 150): 2,
        (150.1, 180): 3,
        (180.1, 240): 4,
        (240.1, float('inf')): 5,
    }
    for (low, high), label in ranges.items():
        if low < value <= high:
            return label
    return -1
```

**Kod definiuje dwie funkcje:**

1. `calculate_air_quality(row)` - Określa indeks jakości powietrza dla wiersza danych, analizując poziomy zanieczyszczeń (`NO2`, `SO2`, `PM25`, `PM10`, `O3`) za pomocą wcześniej zdefiniowanych funkcji klasyfikujących. Zwraca najwyższą wartość indeksu spośród dostępnych zanieczyszczeń lub -1, jeśli dane są niedostępne.

2. `get_air_quality_style(quality)` - Mapuje indeks jakości powietrza (0–5) na odpowiadającą mu ikonę i kolor do celów wizualizacji. W przypadku nieznanej jakości przypisywana jest ikona z pytajnikiem i kolor fioletowy.

```python
def calculate_air_quality(row):
    qualities = []

    if 'NO2' in row and not pd.isna((row['NO2'].item())):
        qualities.append(NO2((row['NO2']).item()))
    if 'SO2' in row and not pd.isna((row['SO2'].item())):
        qualities.append(SO2((row['SO2']).item()))
    if 'PM25' in row and not pd.isna((row['PM25'].item())):
        qualities.append(PM25((row['PM25']).item()))
    if 'PM10' in row and not pd.isna((row['PM10'].item())):
        qualities.append(PM10((row['PM10']).item()))
    if 'O3' in row and not pd.isna((row['O3'].item())):
        qualities.append(O3((row['O3']).item()))
    if len(qualities) == 0:
        qualities.append(-1)

    return max(qualities)

def get_air_quality_style(quality): 
    if quality == 0:
        return ('fa-solid fa-face-grin-beam', 'green')
    if quality == 1:
        return ('fa-solid fa-face-smile-beam', 'lightgreen')
    if quality == 2:
        return ('fa-solid fa-face-meh', 'yellow')
    if quality == 3:
        return ('fa-solid fa-face-frown', 'orange')
    if quality == 4:
        return ('fa-solid fa-face-sad-tear', 'red')
    if quality == 5:
        return ('fa-solid fa-face-angry', 'brown')
    else:
        return ('fa-solid fa-question', 'purple')
```

**Kod oblicza roczne i łączone średnie składników jakości powietrza (NO2, SO2, PM10, PM25, O3) na podstawie plików CSV dla lat 2017-2023 i zapisuje je w DataFrame’ach.
```python
# Funkcja do wczytania danych, transpozycji i obliczenia średnich
def calculate_averages(file_path, element_name):
    df = pd.read_csv(file_path, sep=',')
    df = df.iloc[:, 1:]
    new_columns = [col.split('-')[0] for col in df.columns]
    df.columns = new_columns
    df = df.apply(pd.to_numeric, errors='coerce')
    return df.mean(axis=0).rename(element_name).round(1)

# Lista plików do przetworzenia
files_2017 = {
    'NO2': 'data/2017_NO2_1g.csv',
    'SO2': 'data/2017_SO2_1g.csv',
    'PM10': 'data/2017_PM10_1g.csv',
    'PM25': 'data/2017_PM25_1g.csv',
    'O3': 'data/2017_O3_1g.csv'
}

files_2018 = {
    'NO2': 'data/2018_NO2_1g.csv',
    'SO2': 'data/2018_SO2_1g.csv',
    'PM10': 'data/2018_PM10_1g.csv',
    'PM25': 'data/2018_PM25_1g.csv',
    'O3': 'data/2018_O3_1g.csv'
}

files_2019 = {
    'NO2': 'data/2019_NO2_1g.csv',
    'SO2': 'data/2019_SO2_1g.csv',
    'PM10': 'data/2019_PM10_1g.csv',
    'PM25': 'data/2019_PM25_1g.csv',
    'O3': 'data/2019_O3_1g.csv'
}

files_2020 = {
    'NO2': 'data/2020_NO2_1g.csv',
    'SO2': 'data/2020_SO2_1g.csv',
    'PM10': 'data/2020_PM10_1g.csv',
    'PM25': 'data/2020_PM25_1g.csv',
    'O3': 'data/2020_O3_1g.csv'
}

files_2021 = {
    'NO2': 'data/2021_NO2_1g.csv',
    'SO2': 'data/2021_SO2_1g.csv',
    'PM10': 'data/2021_PM10_1g.csv',
    'PM25': 'data/2021_PM25_1g.csv',
    'O3': 'data/2021_O3_1g.csv'
}

files_2022 = {
    'NO2': 'data/2022_NO2_1g.csv',
    'SO2': 'data/2022_SO2_1g.csv',
    'PM10': 'data/2022_PM10_1g.csv',
    'PM25': 'data/2022_PM25_1g.csv',
    'O3': 'data/2022_O3_1g.csv'
}

files_2023 = {
    'NO2': 'data/2023_NO2_1g.csv',
    'SO2': 'data/2023_SO2_1g.csv',
    'PM10': 'data/2023_PM10_1g.csv',
    'PM25': 'data/2023_PM25_1g.csv',
    'O3': 'data/2023_O3_1g.csv'
}

files_joint = {
    'NO2': 'data/NO2_1g_joint_2017-2023.csv',
    'SO2': 'data/SO2_1g_joint_2017-2023.csv',
    'PM10': 'data/PM10_1g_joint_2017-2023.csv',
    'PM25': 'data/PM25_1g_joint_2017-2023.csv',
    'O3': 'data/O3_1g_joint_2017-2023.csv'
}

# Przechowuje średnie dla każdego punktu w latach 2017-2023
averages_2017 = pd.DataFrame()
averages_2018 = pd.DataFrame()
averages_2019 = pd.DataFrame()
averages_2020 = pd.DataFrame()
averages_2021 = pd.DataFrame()
averages_2022 = pd.DataFrame()
averages_2023 = pd.DataFrame()
averages_joint = pd.DataFrame()

averages = []

# Oblicz średnie dla każdego pliku
#2017
for element, file in files_2017.items():
    element_averages = calculate_averages(file, element)  # Średnie dla danego pierwiastka
    averages_2017[element] = element_averages  # Dodaj jako kolumnę

averages_2017.reset_index(inplace=True)  # Zresetuj indeks, aby "punkty" były kolumną
averages_2017.rename(columns={'index': 'Point'}, inplace=True)  # Nazwij kolumnę z punktami
averages.append([averages_2017, '2017'])
#2018
for element, file in files_2018.items():
    element_averages = calculate_averages(file, element)
    averages_2018[element] = element_averages

averages_2018.reset_index(inplace=True)
averages_2018.rename(columns={'index': 'Point'}, inplace=True)
averages.append([averages_2018, '2018'])
#2019
for element, file in files_2019.items():
    element_averages = calculate_averages(file, element)
    averages_2019[element] = element_averages

averages_2019.reset_index(inplace=True)
averages_2019.rename(columns={'index': 'Point'}, inplace=True)
averages.append([averages_2019, '2019'])
#2020
for element, file in files_2020.items():
    element_averages = calculate_averages(file, element)
    averages_2020[element] = element_averages

averages_2020.reset_index(inplace=True)
averages_2020.rename(columns={'index': 'Point'}, inplace=True)
averages.append([averages_2020, '2020'])
#2021
for element, file in files_2021.items():
    element_averages = calculate_averages(file, element)
    averages_2021[element] = element_averages

averages_2021.reset_index(inplace=True)
averages_2021.rename(columns={'index': 'Point'}, inplace=True)
averages.append([averages_2021, '2021'])
#2022
for element, file in files_2022.items():
    element_averages = calculate_averages(file, element)
    averages_2022[element] = element_averages

averages_2022.reset_index(inplace=True)
averages_2022.rename(columns={'index': 'Point'}, inplace=True)
averages.append([averages_2022, '2022'])
#2023
for element, file in files_2023.items():
    element_averages = calculate_averages(file, element)
    averages_2023[element] = element_averages

averages_2023.reset_index(inplace=True)
averages_2023.rename(columns={'index': 'Point'}, inplace=True)
averages.append([averages_2023, '2023'])
#Wszystkie lata
for element, file in files_joint.items():
    element_averages = calculate_averages(file, element)
    averages_joint[element] = element_averages

averages_joint.reset_index(inplace=True)
averages_joint.rename(columns={'index': 'Point'}, inplace=True)
averages.append([averages_joint, 'joint'])
```

Animowana warstwa czasowa `TimestampedGeoJson` umożliwia wizualizację zmian w ramach osi czasu. Gotowa mapa jest zapisywana w pliku `air_quality_timeline_map.html`.**

```python
def svg_to_base64(file_path):
    """Konwertuje plik SVG na ciąg Base64."""
    with open(file_path, "rb") as f:
        encoded = base64.b64encode(f.read()).decode("utf-8")
    return f"data:image/svg+xml;base64,{encoded}"

def get_air_quality_style_url(quality):
    if quality == 0:
        return svg_to_base64("icons/face-grin-beam-solid.svg")
    if quality == 1:
        return svg_to_base64("icons/face-smile-beam-solid.svg")
    if quality == 2:
        return svg_to_base64("icons/face-meh-solid.svg")
    if quality == 3:
        return svg_to_base64("icons/face-frown-solid.svg")
    if quality == 4:
        return svg_to_base64("icons/face-sad-tear-solid.svg")
    if quality == 5:
        return svg_to_base64("icons/face-angry-solid.svg")
    else:
        return svg_to_base64("icons/circle-question-solid.svg")

def convert_to_geojson(qualities, current_date):
    features = []
    for quality in qualities:
        station_id, coords, air_quality = quality
        icon = get_air_quality_style_url(air_quality)
        popup_text = f"Stacja: {station_id}<br>Jakość powietrza: {quality_map.get(air_quality)}"
        
        features.append({
            "type": "Feature",
            "geometry": {
                "type": "Point",
                "coordinates": coords  # [long, lat]
            },
            'properties': {
                'time': current_date.strftime('%d/%m/%Y'),
                'icon': 'marker',
                'iconstyle': {
                    "iconUrl": icon,
                    'iconSize': [20, 20],
                },
                'popup': popup_text,
                'station': station_id,
                'air_quality': air_quality
            }
        })
    return {"type": "FeatureCollection", "features": features}

m = folium.Map(location=[51.77513554189464, 19.455341364192083], zoom_start=6, tiles='CartoDB positron')

# Przetwarzanie danych
dataframes = []

for element, file_path in files_joint.items():
    df = pd.read_csv(file_path, sep=',')
    dataframes.append([element, df])

# Ustawienie daty początkowej i końcowej
start_date = datetime(2017, 1, 1, 12, 0)
end_date = datetime(2024, 1, 1, 1, 0)

# Ustawienie odstępu godzinowego
interval = timedelta(hours=120)

# Lista do zbierania danych dla całego okresu
all_geojson_data = []

# Przemieszczanie się od start_date do end_date w odstępach godzinowych
current_date = start_date
while current_date < end_date:
    data = defaultdict(list)
    qualities = []
    for dataframe in dataframes:
        filtered_df = dataframe[1][dataframe[1]['Time'] == current_date.strftime('%Y-%m-%dT%H:%M:%S')]
        new_columns = [col.split('-')[0] for col in filtered_df.columns]
        filtered_df.columns = new_columns

        for _, point in filtered_df.iloc[:, 1:].T.iterrows():
            data[point.name].append([dataframe[0], point.iloc[0]])

        max_len = max(len(lst) for lst in data.values())
        for key in data:
            while len(data[key]) < max_len:
                data[key].append(None)
    indexes = pd.DataFrame(dict(data))
    
    for _, i in indexes.T.iterrows():
        # Konwersja danych do słownika
        pollution_data = {item[0]: float(item[1]) if item[1] != 'nan' else None for item in i.dropna()}
        
        # Tworzenie serii Pandas ze słownika
        pollution_series = pd.Series(pollution_data)
    
        # Wywołanie funkcji calculate_air_quality
        try:
            air_quality = calculate_air_quality(pollution_series)
            
            station_data = gdf3[gdf3['StationID'] == i.name]
            if not station_data.empty:  
        
                lat = station_data['lat'].iloc[0]
                long = station_data['long'].iloc[0]
            
                qualities.append([i.name, [long, lat], air_quality])
        except Exception as e:
            print(f"Błąd podczas obliczania jakości powietrza dla {i.name}: {e}")
    
    # Konwersja do GeoJSON
    geojson_data = convert_to_geojson(qualities, current_date)
    all_geojson_data.append(geojson_data)
    
    current_date += interval

# Dodanie warstwy czasowej
TimestampedGeoJson(
    data={
        "type": "FeatureCollection",
        "features": [feature for geojson in all_geojson_data for feature in geojson['features']]
    },
    transition_time=2000,
    period="P5D",
    duration = 'P5D',
    add_last_point=True,
    auto_play=False,
    loop=False
).add_to(m)

# Zapis mapy do pliku HTML
m.save(f'air_quality_timeline_map.html')

#display(m)
```

Jakość powietrza w Polsce wykazuje znaczną sezonowość. Zimą, poziom zanieczyszczeń, zwłaszcza pyłu zawieszonego (PM10 i PM2.5), znacznie wzrasta, co jest spowodowane intensywnym ogrzewaniem budynków. Latem jakość powietrza zwykle się poprawia, choć mogą wystąpić epizody z większą ilością zanieczyszczeń. Wiele małych miejscowości oraz obszarów wiejskich nadal zmaga się z wysokim poziomem zanieczyszczeń, głównie z powodu emisji z kotłów na paliwa stałe oraz spalania śmieci.<br>
## Podsumowanie
W latach 2017-2023 jakość powietrza w Polsce była tematem licznych analiz i działań mających na celu jej poprawę. Mimo zauważalnej poprawy w niektórych obszarach, nadal istnieją wyzwania, które wymagają dalszych działań, zwłaszcza w małych miejscowościach i na terenach wiejskich oraz w największych metropoliach, szczególnie na Śląsku i w Małopolsce. Kluczowym elementem przyszłych działań powinna być dalsza edukacja społeczeństwa, wsparcie inicjatyw obywatelskich oraz kontynuacja polityki zmierzającej do ograniczenia emisji zanieczyszczeń, poprzez działania takie jak:<br>
1. Kontynuowanie i rozszerzanie programów wsparcia dla osób prywatnych i samorządów w zakresie wymiany starych pieców węglowych i poprawy efektywności energetycznej budynków.
2. Zwiększenie inwestycji w transport publiczny, elektromobilność oraz infrastrukturę rowerową, co mogłoby zredukować emisje spalin z transportu.
3. Wprowadzenie bardziej restrykcyjnych norm emisji spalin oraz monitorowanie jakości powietrza w czasie rzeczywistym w mniejszych miejscowościach.
